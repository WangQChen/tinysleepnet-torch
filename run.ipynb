```python
import matplotlib.pyplot as plt
import pandas as pd
import re
```

```python
log_path = "logs/run_20250609_212215/train.log"

train_loss, train_acc, train_f1 = [], [], []
val_loss, val_acc, val_f1 = [], [], []

with open(log_path, 'r') as f:
    for line in f:
        if 'Train Loss' in line:
            match = re.search(r"Train Loss:\s*([\d.]+), Acc:\s*([\d.]+), F1:\s*([\d.]+)", line)
            if match:
                train_loss.append(float(match.group(1)))
                train_acc.append(float(match.group(2)))
                train_f1.append(float(match.group(3)))
        elif 'Valid Loss' in line:
            match = re.search(r"Valid Loss:\s*([\d.]+), Acc:\s*([\d.]+), F1:\s*([\d.]+)", line)
            if match:
                val_loss.append(float(match.group(1)))
                val_acc.append(float(match.group(2)))
                val_f1.append(float(match.group(3)))


plt.figure(figsize=(10, 5))
plt.plot(train_loss, label='Train Loss')
plt.plot(val_loss, label='Val Loss')
plt.title('Loss over Epochs')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(train_acc, label='Train Acc', marker='o')
plt.plot(val_acc, label='Val Acc', marker='x')
plt.title('Accuracy over Epochs')
plt.ylim(0.6, 0.85)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(train_f1, label='Train F1', marker='o')
plt.plot(val_f1, label='Val F1', marker='x')
plt.title('F1 Score over Epochs')
plt.legend()
plt.show()

```
